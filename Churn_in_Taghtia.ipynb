**VIII.DATA PREPROCESSING**

```python
data = data.drop(['Customer_Name'], axis = 1)
data.head()
```

```python
def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python
data = data.apply(lambda x: object_to_int(x))
data.head(10)
```

```python
plt.figure(figsize=(10,5))
print(data.corr()['Churn'].sort_values(ascending = False))
```

```python
X = data.drop(columns = ['Churn'])
y = data['Churn'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state = 40, stratify=y)
```

```python
def distplot(feature, frame, color='springgreen'):
    plt.figure(figsize=(8,3))
    plt.title("Distribution for {}".format(feature))
    ax = sns.distplot(frame[feature], color= color)
num_cols = ["Tenure", 'Monthly_Charges', 'Total_Charges']
for feat in num_cols: distplot(feat, data)
```

* tant donn que les caractristiques numriques sont rparties sur diffrentes plages de valeurs, j'utiliserai un scalaire standard pour les rduire  la mme plage.

*Standardizing numeric attributes*

```python
data_std = pd.DataFrame(StandardScaler().fit_transform(data[num_cols].astype('float64')),
                       columns=num_cols)
num_cols = ["Tenure", 'Monthly_Charges', 'Total_Charges']
for feat in num_cols: distplot(feat, data_std, color='mediumblue')
```

```python
# Divisez les colonnes en 3 catgories, une pour la normalisation, une pour l'encodage des tiquettes et une pour l'encodage  chaud

cat_cols_ohe =['Payment_Method', 'Contract', 'Internet_Service'] # those that need one-hot encoding
cat_cols_le = list(set(X_train.columns)- set(num_cols) - set(cat_cols_ohe)) #those that need label encoding
```

```python
scaler= StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

# IV.1.SVM





```python
svc_model = SVC(random_state = 1)
svc_model.fit(X_train,y_train)
predict_y = svc_model.predict(X_test)
accuracy_svc = svc_model.score(X_test,y_test)
print("SVM accuracy is :",accuracy_svc)
```

```python
print(classification_report(y_test, predict_y))
```

# IV.2.Logistic Regression

```python
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
accuracy_lr = lr_model.score(X_test,y_test)
print("Logistic Regression accuracy is :",accuracy_lr)
```

```python
lr_pred= lr_model.predict(X_test)
report = classification_report(y_test,lr_pred)
print(report)
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test, lr_pred),
                annot=True,fmt = "d",cmap='Set3',linewidths=3)
    
plt.title("Logistic Regression Confusion Matrix",fontsize=14)
plt.show()
```

```python
y_pred_prob = lr_model.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot([0, 1], [0, 1], 'k--' )
plt.plot(fpr, tpr, label='Logistic Regression',color = "r")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve',fontsize=16)
plt.show();
```

# IV.6.VotingClassifier

* Prdisons maintenant le modle final bas sur la plus grande majorit des votes et vrifions son score.

```python
from sklearn.ensemble import VotingClassifier
clf1 = GradientBoostingClassifier()
clf2 = LogisticRegression()
clf3 = AdaBoostClassifier()
eclf1 = VotingClassifier(estimators=[('gbc', clf1), ('lr', clf2), ('abc', clf3)], voting='soft')
eclf1.fit(X_train, y_train)
predictions = eclf1.predict(X_test)
print("Final Accuracy Score ")
print(accuracy_score(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test, predictions),
                annot=True,fmt = "d",cmap='flag',linewidths=3)
    
plt.title("Final Confusion Matrix",fontsize=14)
plt.show()
```
